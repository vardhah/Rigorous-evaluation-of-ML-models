```python
import pandas as pd 
import numpy as np
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
import tensorflow as tf
```

# Loading and Visualising Training Data 

```python
data_training=pd.read_csv('Data_training.csv')
failed_training = data_training[data_training['Stop_Distance'] <= 0]
passed_training = data_training[data_training['Stop_Distance'] > 0]
print('Total number of episodes:',data_training['Episode'].shape[0])

print('Size of crashes :',failed_training.shape[0])  
print('Size of success :',passed_training.shape[0]) 
print("Percentage of crash cases:{:.2f}".format(failed_training.shape[0]*100/(failed_training.shape[0]+passed_training.shape[0])),"%")
print("Percentage of not crashed cases:{:.2f}".format(passed_training.shape[0]*100/(failed_training.shape[0]+passed_training.shape[0])),"%")
data_training.head()
```

```python
fig,ax=plt.subplots()
plt.scatter(data_training['Episode'],data_training['friction'])
ax.grid(True) 
plt.show()
data_training.loc[data_training['Stop_Distance'] <= 0, 'test'] = int(1)   # 1 indicates failure
data_training.loc[data_training['Stop_Distance'] > 0, 'test'] = int(0)    # 0 indicated pass

print('Total number of episodes:',data_training['Episode'].shape[0])
start=8000
end=9998
fig, ax = plt.subplots()
episode=np.array(data_training['Episode'].values.tolist()[start:end])
test=np.array(data_training['test'].values.tolist()[start:end])
speed=np.array(data_training['NN_start_speed'].values.tolist()[start:end])

indices = test == 0
print('Number of Pass:',indices.sum())
print('Number of fail:',indices.shape[0]-indices.sum())

# normal values
ax.scatter(episode[~indices], speed[~indices], c = 'red',label='fail')
# bad values
ax.scatter(episode[indices], speed[indices], c = 'grey',label='Pass')
#ax.legend()
#ax.set_ylim([0, 45])
plt.xlabel('episode')
plt.ylabel('speed (m/s)')
ax.grid(True) 
plt.show()
```

# Training for AVF

```python
#data for AVF training:
iteration_start=8000
iteration_end=9998
data=data_training[["Kick_Speed","friction","test"]][iteration_start:iteration_end]
data.head()
print(data)
```

```python
#data normalisation for AVF training:
std=np.std(data[["Kick_Speed","friction"]].values,axis=0)
mean=np.mean(data[["Kick_Speed","friction"]].values,axis=0)
np.savez('stdmean.npz',std=std, mean=mean)
data[[ "Kick_Speed","friction"]]=np.divide(np.subtract(data[[ "Kick_Speed","friction"]],mean),std)
print('std is:',std.shape)
print('Mean is:',mean)
data.head()
```

```python
#AVF Training 
X=data[["Kick_Speed","friction"]].values
print(X.shape)
Y=data["test"]
print(Y.shape)
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split

#test train split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
print(X[1,:].shape)

#callback
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('acc')>0.97):
      print("\nReached 97% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = myCallback()

# Model

model = tf.keras.models.Sequential([tf.keras.layers.InputLayer(input_shape=(2,)), 
                                    tf.keras.layers.Dense(64, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(32, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(1, activation=tf.nn.sigmoid )
])
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=500, batch_size=8, callbacks=[callbacks]) ;
```

```python
results = model.evaluate(x_test, y_test, batch_size=8)
print('test loss, test acc:', results)
```
